#correlation with happiness score

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical columns for correlation
df1_numeric = df1.select_dtypes(include=['number'])

# Calculate and display the correlation matrix
correlation_matrix = df1_numeric.corr()
display(correlation_matrix['Happiness Score'].sort_values(ascending=False).iloc[1:])
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of Happiness Score vs. Economy (GDP per Capita) for df1
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df1, x='Economy (GDP per Capita)', y='Happiness Score')
plt.title('Happiness Score vs. Economy (GDP per Capita) (df1)')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()

# Scatter plot of Happiness Score vs. Dystopia Residual for df1
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df1, x='Dystopia Residual', y='Happiness Score')
plt.title('Happiness Score vs. Dystopia Residual (df1)')
plt.xlabel('Dystopia Residual')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()
```
